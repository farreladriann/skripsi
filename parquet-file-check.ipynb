```python
import os
from shared import DATA_FILES
import pyarrow.parquet as pq

BASE_DIR = os.path.abspath('.')
DATA_ROOT = os.path.join(BASE_DIR, 'DataGPS_parquet')

print(f'BASE_DIR  : {BASE_DIR}')
print(f'DATA_ROOT : {DATA_ROOT}')
print(f'File data : {len(DATA_FILES)} bulan')
```

```python
def cek_struktur(path: str) -> dict:
    """Schema, tipe data, jumlah baris & kolom, metadata file."""
 
    pf = pq.ParquetFile(path)
    meta = pf.metadata
    schema = pf.schema_arrow
 
    info = {
        "path": path,
        "ukuran_mb": os.path.getsize(path) / 1e6,
        "n_row": meta.num_rows,
        "n_col": len(schema),
        "n_row_groups": meta.num_row_groups,
        "kompresi": meta.row_group(0).column(0).compression,
        "created_by": meta.created_by,
        "kolom": {field.name: str(field.type) for field in schema},
    }
```

```python
file_path = os.path.join(DATA_ROOT, DATA_FILES[0][1])
cek_struktur(file_path)
```
